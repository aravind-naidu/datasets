## Data

```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('')
df.head()
```

```
df.shape
```

```
language_list = df['language'].unique()
language_list
```

```
df['proj_id'].nunique()
```

```
df['file_id'].nunique()
```

```
def language_count(lang):
  count = (df['language']==lang).sum()
  print(count)
```

```
language_count('JavaScript')
```

```
lan = []
for i in range(len(language_list)):
  count = (df['language']==language_list[i]).sum()
  lan.append(count)

lan
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))

plt.bar(language_list,lan)
plt.show()
```

```
for i in range(len(language_list)):
  print(f'{language_list[i]:} count is      {lan[i]}')
```

```
plt.figure(figsize=(15,10))
plt.pie(lan, explode=None, labels=language_list,pctdistance=0.6, autopct='%1.1f%%',shadow=True, startangle=90)
```
